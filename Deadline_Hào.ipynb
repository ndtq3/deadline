```python
<!DOCTYPE html>
<html lang="vi">
<head>
<meta charset="UTF-8">
<title>Countdown Widget</title>

<style>
  body {
    background: #020617;
    margin: 0;
    padding: 40px;
    display: flex;
    justify-content: center;
  }

  .widget {
    width: 520px;              /* khung to hơn */
    padding: 32px 36px;
    border: 2px solid #facc15;
    border-radius: 20px;
    font-family: Arial, sans-serif;
    text-align: center;
    background: #020617;
  }

  .deadline {
    font-size: 18px;
    color: #e5e7eb;
    margin-bottom: 12px;
  }

  .person {
    font-size: 20px;
    color: #ef4444;
    margin-bottom: 20px;
  }

  #timer {
    font-size: 48px;           /* số đếm RẤT TO */
    font-weight: bold;
    color: #facc15;
  }

  .expired {
    color: #ef4444;
    font-size: 46px;
  }
</style>

</head>

<body>

<div class="widget">
  <div class="deadline">
    Hạn chót: 20:00 – 31/01/2026 (GMT+7)
  </div>

  <div class="person" id="personName">
    Bùi Anh Hào – Ban Kế Hoạch
  </div>

  <div id="timer">Đang tính…</div>
</div>

<script>
  // ===== CẤU HÌNH =====
  const personName = "Bùi Anh Hào – Bản Kế Hoạch";

  // Hạn chót: 20:00 VN = 13:00 UTC
  const deadlineUTC = Date.UTC(2026, 0, 31, 13, 0, 0);

  document.getElementById("personName").innerText = personName;

  function updateTimer() {
    const nowUTC = new Date().getTime();
    let diff = deadlineUTC - nowUTC;

    if (diff <= 0) {
      document.getElementById("timer").innerHTML =
        "<span class='expired'>⛔ ĐÃ HẾT HẠN</span>";
      return;
    }

    let days = Math.floor(diff / (1000 * 60 * 60 * 24));
    let hours = Math.floor((diff / (1000 * 60 * 60)) % 24);
    let minutes = Math.floor((diff / (1000 * 60)) % 60);
    let seconds = Math.floor((diff / 1000) % 60);

    document.getElementById("timer").innerText =
      days + " ngày " +
      String(hours).padStart(2, "0") + ":" +
      String(minutes).padStart(2, "0") + ":" +
      String(seconds).padStart(2, "0");
  }

  setInterval(updateTimer, 1000);
  updateTimer();
</script>

</body>
</html>

```
